```python
import pandas as pd
```

```python
data = pd.read_csv('data/hexagon.csv')
data
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8, 6))
sns.scatterplot(data=data, x='x', y='y', hue='c', palette='viridis', s=100)
plt.title('Wizualizacja punktÃ³w')
plt.legend(title='Kolor (c)')
plt.show()
```

```python
data2 = pd.read_csv('data/cube.csv')
data2
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

scatter = ax.scatter(
    data2['x'],
    data2['y'],
    data2['z'],
    c=data2['c'],
    cmap='viridis',   # paleta kolorÃ³w
    s=50              # wielkoÅ›Ä‡ punktu
)

fig.colorbar(scatter, ax=ax, label='Kolor (c)')

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_title('3D Wizualizacja punktÃ³w')

plt.show()
```
